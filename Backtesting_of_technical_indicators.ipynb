```python
from fastquant import get_stock_data
df = get_stock_data("MSFT", "2006-01-01", "2015-01-01")
print(df.tail())
```

```python
from fastquant import backtest
backtest('smac', df, fast_period=50, slow_period=200)
```

```python
backtest('rsi', df, rsi_period=14, rsi_upper=70, rsi_lower=30)
```

```python
backtest('emac', df, fast_period=50, slow_period=200)
```

```python
backtest('bbands', df, period=20, devfactor=2.0)
```

```python
backtest('buynhold', df)
```
